```python
import requests
from geopy.geocoders import Nominatim
from IPython.display import display, HTML
import ipywidgets as widgets
API_KEY = 'a12a80e7b5454ea82ebf6db1a40ff2e0'
def get_current_weather(city):
    """Fetches current weather data for a given city."""
    url = f'http://api.openweathermap.org/data/2.5/weather?q={city}&appid={API_KEY}&units=metric'
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error fetching weather data: {response.status_code} - {response.json()}")
        return None
def get_forecast(city):
    """Fetches 5-day weather forecast data for a given city."""
    url = f'http://api.openweathermap.org/data/2.5/forecast?q={city}&appid={API_KEY}&units=metric'
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error fetching forecast data: {response.status_code} - {response.json()}")
        return None
def get_weather_by_location():
    """Fetches weather data based on the user's current location."""
    geolocator = Nominatim(user_agent="weather_app")
    location = geolocator.geocode("Your location")  # Use your city or 'Your location' placeholder
    if location:
        return get_current_weather(location.address)
    else:
        print("Could not determine your location.")
        return None
def display_weather(data):
    """Displays weather data with icons."""
    if data:
        city_name = data['name']
        temp = data['main']['temp']
        weather_desc = data['weather'][0]['description'].title()
        humidity = data['main']['humidity']
        wind_speed = data['wind']['speed']
        icon_code = data['weather'][0]['icon']
        icon_url = f"http://openweathermap.org/img/wn/{icon_code}.png"

        weather_info = f"""
        <h3>Weather in {city_name}</h3>
        <p><img src="{icon_url}" alt="Weather Icon"> {weather_desc}</p>
        <p>Temperature: {temp}°C</p>
        <p>Humidity: {humidity}%</p>
        <p>Wind Speed: {wind_speed} m/s</p>
        """
        display(HTML(weather_info))
def display_forecast(data):
    """Displays 5-day weather forecast data."""
    if data:
        forecast_info = "<h3>5-Day Weather Forecast</h3>"
        for forecast in data['list']:
            date = forecast['dt_txt']
            temp = forecast['main']['temp']
            weather_desc = forecast['weather'][0]['description'].title()
            icon_code = forecast['weather'][0]['icon']
            icon_url = f"http://openweathermap.org/img/wn/{icon_code}.png"

            forecast_info += f"""
            <p><b>{date}</b></p>
            <p><img src="{icon_url}" alt="Weather Icon"> {weather_desc}, {temp}°C</p>
            <hr>
            """
        display(HTML(forecast_info))
city_input = widgets.Text(value='', description='Enter city:')
display_button = widgets.Button(description='Get Weather')
def on_button_click(b):
    city = city_input.value
    if city:
        weather_data = get_current_weather(city)
        display_weather(weather_data)
        
        forecast_data = get_forecast(city)
        display_forecast(forecast_data)
    else:
        display(HTML("<p style='color: red;'>Please enter a city name.</p>"))

display_button.on_click(on_button_click)

display(city_input, display_button)
```
